```python
#Log-in screen
 
root = Tk()
root.title('Piano Sales Management System')
root.geometry('450x200')
root.resizable(False, False)


def login():
    if (username_entry.get() == data['user']) & (password_entry.get() == data['password']):
	    root.destroy()
    else:
        messagebox.showerror(title='Sai thÃ´ng tin Ä‘Äƒng nháº­p', 
                             message='Vui lÃ²ng kiá»ƒm tra láº¡i username vÃ  password')
          
# Labels and entries for username and password
username_label = Label(root, text='Username', font=('Andalus', 15, 'bold'), fg='brown')
password_label = Label(root, text='Password', font=('Andalus', 15, 'bold'),fg='brown')

username_entry_var = StringVar()
password_entry_var = StringVar()

username_entry = Entry(root, textvariable=username_entry_var, 
                                font=('Andalus', 15))
password_entry = Entry(root, textvariable=password_entry_var, show='*', 
                                        font=('Andalus', 15))
btn_login = Button(root, text='LOGIN', activebackground='SteelBlue1', activeforeground='white',
                                fg='brown', bg='#F0F8FF', font=('Andalus', 15, 'bold'),
                                command=login)

username_label.place(x=20, y=20)
password_label.place(x=20, y=60)
username_entry.place(x=160, y=20, width=200)
password_entry.place(x=160, y=60, width=200)
btn_login.place(x=160, y=100, width=200)

# Opening JSON file
f = open('id.json')
# returns JSON object as a dictionary
data = json.load(f)
# Closing file
f.close()

root.mainloop()
```

```python
from tkinter import *
from tkinter import ttk
from tkinter.messagebox import showinfo
import json

root = Tk()
root.title('VIEW ALL PIANO')
root.geometry('700x700')



Canvas1 = Canvas(root) 
Canvas1.config(bg="old lace")
Canvas1.pack(expand=True,fill=BOTH)

headingFrame1 = Frame(root,bg='old lace',bd=5)
headingFrame1.place(relx=0.25,rely=0.1,relwidth=0.5,relheight=0.13)

headingLabel = Label(headingFrame1, text="View Products", fg='black', font = ('Courier',15,'bold'))
headingLabel.place(relx=0,rely=0, relwidth=1, relheight=1)

# labelFrame = Frame(root)
# labelFrame.place(relx=0.1,rely=0.3,relwidth=0.8,relheight=0.5)
# y = 0.25
# define columns
columns = ("Product brand", "Product price", "Date")

tree = ttk.Treeview(root, columns=columns, show='headings')

# define headings
tree.heading("Product brand", text="Product brand")
tree.heading("Product price", text="Product price")
tree.heading("Date", text="Date")


# Opening JSON file
f = open('product.json')
# returns JSON object as a dictionary
data = json.load(f)
# Closing file
f.close()
count=0
for prod in data['piano']:
    tree.insert(parent='', index="end", iid=count, text="", 
                values=[prod["Product brand"], prod["Product price"], prod["Date"]])
    count+=1
# # add data to the treeview
# for prod in data["piano"]:
#     tree.insert('', tk.END, values=[prod["Product brand"], prod["Product price"], prod["Date"]])



tree.place(relx=0.07,rely=0.3)

# add a scrollbar
scrollbar = ttk.Scrollbar(root, orient=VERTICAL, command=tree.yview)
tree.configure(yscroll=scrollbar.set)
scrollbar.place(relx=0.93,rely=0.3,relheight=0.325)

# run the app
root.mainloop()
```
